## Question 3 ‚Äî Write Functions *(20 points)*

```python
# üìù [Q3a] Sum of even numbers
def even_sum(lst):
    return sum(x for x in lst if x % 2 == 0)

# Tests:
assert even_sum([1, 2, 3, 4, 5, 6]) == 12
assert even_sum([1, 3, 5]) == 0
assert even_sum([2, 4, 6]) == 12
assert even_sum([]) == 0
print("‚úÖ even_sum: 4/4 tests passed")
```

```python
# üìù [Q3b] Count vowels (case-insensitive)
def count_vowels(text):
    return sum(1 for c in text.lower() if c in 'aeiou')

# Tests:
assert count_vowels("Hello World") == 3
assert count_vowels("aeiou") == 5
assert count_vowels("AEIOU") == 5
assert count_vowels("xyz") == 0
print("‚úÖ count_vowels: 4/4 tests passed")
```
